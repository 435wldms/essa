**앤스콤 데이터 집합 모두 사용해 그래프 만들기**

* matplotlib 라이브러리로 그래프 그리기
>
1) 전체 그래프가 위치할 기본 틀 만들기

2) 그래프를 그려 넣을 그래프 격자를 만들기
>
3) 격자에 그래프를 하나씩 추가합니다. 격자에 그래프가 추가되는 순서는 왼 -> 오
>
4) 격자의 첫번째 행이 꽉 차면 두 번째 행에 그래프를 그려 넣기

```python
# 한 번에 4개의 그래프 그리기

# 1. 앤스콤 데이터프레임의 dataset 열의 값이 I, II, III, IV 인 것을 불린 추출하여 dataset_1,2,3,4에 저장
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

```python
# 2. 그래프 격자가 위치할 기본 틀 만들기
fig = plt.figure()
# 3. add_subplot 메서드로 그래프 격자 그리기(첫 번째 인자는 기본 틀의 행 크기, 두 번째 인자는 기본 틀의 열 크기)
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
```

```python
# 4. plot 메서드에 데이터를 전달하여 그래프 그리기
axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

# 5. 각각의 그래프를 구분하기 위해 그래프 격자에 제목 추가
axes1.set_title('dataset_1')
axes2.set_title('dataset_1')
axes3.set_title('dataset_1')
axes4.set_title('dataset_1')

fig
```

```python
#6. 기본 틀에 제목 추가
fig.suptitle('Anscombe Data')
fig
```

```python
# 7. 그래프의 이름과 숫자가 겹치지 않게 하기
fig.tight_layout()
fig
```

# 04-2 matplotlib 라이브러리 자유자재로 사용하기

**기초 그래프 그리기**
>
1) 히스토그램

```python
# 1. tips 데이터 집합 불러와 변수에 저장
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

```python
# 2. 기본 틀 마련하고 격자를 구성하기
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
# 3. hist 메서드에 total_bill 열 전달하기(bins 인잣값으로 x축 간격 조정)
axes1.hist(tips['total_bill'], bins =10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

2) 산점도(이변량 그래프)

```python
scatter_plot= plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

3) 박스 그래프(이산형 변수와 연속형 변수 함께 사용)

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],tips[tips['sex']=='Male']['tip']],labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

**다변량 그래프 그리기**

```python
# 산점도 그래프
# 1. 성별을 산점도 그래프에 새 변수로 추가하기 위해서 여성, 남성을 0,1로 바꿈
def recode_sex(sex):
  if sex =='Female':
    return 0
  else:
    return 1
```

```python
# 2. recode_sex 메서드가 반환한 값을 dataframe에 sex_color 변수로 추가
tips['sex_color']=tips['sex'].apply(recode_sex)
```

```python
# 3. table 당 인원 수,성별을 s,c 인잣값으로 그래프에 추가
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5
)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_xlabel('Tip')
```

# 04-3 seaborn 라이브러리 자유자재로 사용하기

단변량 그래프 그리기 - 히스토그램

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
# subplot으로 기본 틀 만들고, distplot에 원하는 열 전달
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
#밀집도 그래프 제외: kde=False
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 밀집도 그래프만 나타내는법(hist=False)
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
# 히스토그램 + 밀집도 + 양탄자 그래프(rug)
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
#count 그래프
ax = plt.subplots()
ax = sns.countplot(x='day',data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

이변량 그래프 그리기

# 04-4 데이터프레임과 시리즈로 그래프 그리기

```python
# histogram
ax = plt.subplots()
ax = tips['total_bill'].plot.hist()

# 투명도 조절(alpha), x축 간격(bins)
fig, ax = plt.subplots()
ax = tips[['total_bill','tip']].plot.hist(alpha=0.5, bins=20, ax=ax)
```

```python
# 밀집도(kde)
ax = plt.subplots()
ax = tips['tip'].plot.kde()

# 산점도(scatter)
fig, ax = plt.subplots()
ax = tips.plot.scatter(x='total_bill',y='tip',ax=ax)

# 육각 그래프(hexbin)
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', ax=ax)
# gradsize 인자로 육각형 크기 변경 가능
fig, ax = plt.subplots()
ax= tips.plot.hexbin(x='total_bill',y='tip',gridsize=10,ax=ax)

# 상자(box)
fig, ax = plt.subplots()
ax = tips.plot.box(ax=ax)
```

# 04-5 seaborn 라이브러리로 그래프 스타일 설정하기

```python
# white grid 스타일
sns.set_style('whitegrid')
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)

# for 문을 활용해 모든 스타일 하나씩 적용

fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx +1
  with sns.axes_style(style):
    ax = fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips,ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
