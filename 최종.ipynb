# 변수 설명
* index: 구분자
* quality: 품질
* fixed acidity: 산도
* volatile acidity: 휘발성산
* citric acid: 시트르산
* residual sugar: 잔당 - 발효 후 와인 속에 남아있는 당분
* chlorides: 염화물

```python
# One-Hot Encoding
train = pd.get_dummies(train, columns = ["type"], prefix = ["type"])
test = pd.get_dummies(test, columns = ["type"], prefix = ["type"])
```

```python
# x,y 구분 + 의미없는 변수 index 삭제
X_train = train.drop(['index','quality'],axis=1)
y_train = train['quality']

X_train
```

```python
test = test.drop(['index'],axis=1)
```

```python
# numerical_col 뽑기
numerical_col = X_train.select_dtypes(exclude='object').columns.tolist()
numerical_col.remove('type_red')
numerical_col.remove('type_white')
```

```python
numerical_col
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train[numerical_col]=ss.fit_transform(X_train[numerical_col])
test[numerical_col]=ss.fit_transform(test[numerical_col])
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
dt_clf = DecisionTreeClassifier(random_state=156)

params = {
    'max_depth' : [8,12,13,20],
    'min_samples_split' : [16,24],
}
grid_cv = GridSearchCV(dt_clf,param_grid = params, scoring='accuracy',cv=5, verbose=1)
grid_cv.fit(X_train,y_train)

print('GridSearchCV 최고 평균 정확도 수치: {0:.4f}'.format(grid_cv.best_score_))
print('GridSearchCV 최적 하이퍼 파라미터:',grid_cv.best_params_)
```

```python
best_dt_clf = grid_cv.best_estimator_
dt_pred = best_dt_clf.predict(test)
```

```python
submission = pd.read_csv('sample_submission.csv')
```

```python
submission['quality'] = dt_pred
submission
```

```python
submission.to_csv('submission.csv', index=False)
```

-> 0.56699

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = {
    "n_estimators": [100, 200, 300],
    "max_depth": [4, 5, 6],
    "min_samples_leaf": [8, 12, 18],
    "min_samples_split": [8, 16, 20]
}

rfc = RandomForestClassifier()
grid_cv = GridSearchCV(rfc, param_grid = params, cv = 2, n_jobs = -1)
grid_cv.fit(X_train, y_train)
```

```python
print("최적 하이퍼 파라미터: ", grid_cv.best_params_)
print("최고 예측 정확도: {0:.4f}".format(grid_cv.best_score_))
```

```python
# best hyper parameter로 학습하기
best_rf = grid_cv.best_estimator_
best_rf.fit(X_train, y_train)

rfc_pred = best_rf.predict(test)
```

```python
submission = pd.read_csv('sample_submission.csv')
```

```python
submission['quality'] = rfc_pred
submission
```

```python
submission.to_csv('submission1.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ftr_importances_values = best_rf.feature_importances_
ftr_importances = pd.Series(ftr_importances_values, index = X_train.columns)
ftr_top10 = ftr_importances.sort_values(ascending = False)[:10]

plt.figure(figsize = (8, 6))
plt.title("Feature importances Top 10")
sns.barplot(x = ftr_top10, y = ftr_top10.index)
plt.show()
```

-> 0.57

gradient boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

gb_clf = GradientBoostingClassifier()

params ={
    'n_estimators':[100,300,500,700,1000],
    'learning_rate':[0.05,0.1,0.3]
}
grid_cv = GridSearchCV(gb_clf, param_grid=params, cv=3, verbose=1)
grid_cv.fit(X_train,y_train)
print('최적 하이퍼 파라미터:\n',grid_cv.best_params_)
print('최고 예측 정확도:{0:.4f}'.format(grid_cv.best_score_))
```

```python
submission = pd.read_csv('sample_submission.csv')
```

XGBoost

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
le = LabelEncoder()

y_train= le.fit_transform(y_train)
X_tr,X_val,y_tr,y_val=train_test_split(X_train,y_train,test_size=0.3,random_state=121,stratify=y_train)
xgb_wrapper=XGBClassifier(objective='multi:softmax',n_estimators=400,learning_rate=0.05,max_depth=3,eval_metric='mlogloss')
params={'max_depth':[5,7],'min_child_weight':[1,3],'colsample_bytree':[0.5,0.75]}
gridcv=GridSearchCV(xgb_wrapper,param_grid=params,cv=3)
gridcv.fit(X_tr,y_tr,early_stopping_rounds=50,eval_set=[(X_tr,y_tr),(X_val,y_val)])
print('GridSearchCV 최적 파라미터',gridcv.best_params_)
y_train=le.inverse_transform(y_train)
```

```python
best_xgb=gridcv.best_estimator_
best_xgb.fit(X_train,y_train)
xgb_preds=best_xgb.predict(test)+3
```

```python
submission = pd.read_csv('sample_submission.csv')
submission['quality'] = xgb_preds
submission
```

-> 0.662

LightGBM




```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 200으로 감소
lgbm_clf=LGBMClassifier(n_estimators=200,objective='multiclass',metric='multi_logloss')
params={'num_leaves':[32,44,64],'max_depth':[128,140,160],'min_child_samples':[60,100],'subsample':[0.8,1]}
# cv=3 으로 지정
gridcv=GridSearchCV(lgbm_clf,param_grid=params,cv=3,n_jobs=-1)
gridcv.fit(X_train,y_train)
print('GridSearch 최적 파라미터',gridcv.best_params_)
```

```python
print('최고 예측 정확도:{0:.4f}'.format(gridcv.best_score_))
```

```python
# 베스트 파라미터로 학습하기
best_lgbm = gridcv.best_estimator_
best_lgbm.fit(X_train,y_train)
lgbm_preds=best_lgbm.predict(test)
lgbm_preds
```
